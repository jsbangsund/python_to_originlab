```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
import win32com.client
import numpy_to_origin as np2o
```

```python
x = np.arange(-10,10.1,0.1)
y = np.cos(x)
y2 = np.sin(x)
y3 = np.tan(x)

data = np.array((x,y,y2))
header = ['x','cos(x)','sin(x)']
origin,wb,ws = np2o.numpy_to_origin(
    data,column_axis=0,
    long_names=header,
    origin_version=2018,
    worksheet_name='Trig functions',
    workbook_name='data')
```

```python
np2o.createGraph_multiwks(origin,
                          'Trig','Spectra_Wide.otp',
                          os.path.abspath('OriginTemplates'),
                          [ws],[0,0],[1,2],['Sym','Line'])
```
